```python
import requests
from bs4 import BeautifulSoup
```

```python
#URL = "http://poltektedc.ac.id/new/"
URL = "http://poltektedc.ac.id/v6/"
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content, 'html5lib')
```

```python
berita_kampus = soup.findAll("div", {"id": "homeColumns"})
links = []
for berita in berita_kampus:
    link = {}
    link['judul'] = berita.h2.text.replace('n', '')
    link['ringkasan'] = berita.a['href']
    links.append(link)
print(links)
```

```python
page = requests.get(berita.a['href'])
content_article = BeautifulSoup(page.content, 'html.parser')
print(content_article.prettify())
```
